```python
import numpy as np
import pandas as pd
import pickle
from tabulate import tabulate
```

## After 1 iteration with one word

```python
# using test-iters.py with 1 iteration
test_vs_pred_file1 = "one_word/preds_vs_labels_iters1.tsv"
tp1 = pd.read_table(test_vs_pred_file1)
```

```python
tp1_diff = tp1[tp1["same_diff"] == 'DIFF']
tp1_diff
```

```python
print tabulate(tp1_diff[['pred_on_orig', 'pred_on_adv', 'target']], \
         tablefmt='latex', headers='keys')
```

```python
len(tp1_diff) / float(len(tp1))
```
